```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
max_score = -2
for alpha in np.linspace(0.1, 4, 100):
    for beta in np.linspace(0.1, 4, 100):
        x1 = lipid1.reshape(-1, 1)
        y1 = hydro8.reshape(-1, 1)
        y1 = (y1 ** beta) * (x1 ** alpha)
        reg.fit(x1, y1)
        if (max_score < reg.score(x1, y1)):
            max_score = reg.score(x1, y1)
            max_alpha = alpha
            max_beta = beta
print(max_alpha, max_beta)
```
