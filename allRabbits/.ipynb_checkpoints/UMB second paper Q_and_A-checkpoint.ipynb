```python
def BSC_SE(esd, eac, f):
    #print(esd, eac, f)
    L = 1
    q = 2
    a_eff = esd /2
    bsc1 = 185 * L * (q**2) * (a_eff **6) * eac * (f**4)
    bsc1 /= 1+ 2.66 * (f*q*a_eff)**2
    bsc1 *= np.exp(-12.159 * (f**2) * (a_eff **2))
    return bsc1
```

```python
bsc_1  = BSC_SE(esd[0], 10 ** (eac[0]/10), freq)
```
